```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def judge_Euler(G):
    for v in G.nodes(): #各ノードをvとして取り出す
        if G.degree[v] % 2 != 0: #もしノードの次数が偶数でないなら
                return False #Falseを返す
    
    return True #すべてのノードを調査しすべて偶数ならばオイラー閉路といえるのでTrueを返す
```

```python
G = nx.read_edgelist('./test.edgelist1.txt', nodetype=int)
   
nx.draw_networkx(G)
plt.show()

judge_Euler(G)
```

```python
G = nx.read_edgelist('test.edgelist2.txt', nodetype=int)
   
nx.draw_networkx(G)

judge_Euler(G)
```
