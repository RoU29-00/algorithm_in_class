```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.read_edgelist('test.edgelist.txt', nodetype=int)

nx.draw_networkx(G)
plt.show()
```

```python
def my_dfs(G, s, traversed):
    traversed.add(s) #現在位置を到達済みき追加
    print(s) #現在位置を表示
    
    for x in G.neighbors(s): #現在位置の隣接点を取る。
        if not x in traversed: #到達位置に隣接点がない場合、それを初期位置とし、再帰的に繰り返す。
            my_dfs(G, x, traversed)
```

```python
my_dfs(G, 0, set())
```
